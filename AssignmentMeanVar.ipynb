**2. What do you think about the students' parental level of education?**

```python
data['parental level of education'].value_counts().plot.bar()
plt.title('Students parental level of education Count')
plt.xlabel('Level of education')
plt.ylabel('Count')
plt.show()
```

***Inference-Majority of the parents have gone to some College.The number of parents who has an associate degree has the next highest count.It is followed by highschool,some high school,bachelor's degree.The number of parents with the highest level of education among these categories,i..e Master's degree is very low.***

**3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course**

```python
data.groupby('gender').mean()
```

***Inference-For math, males scored the most on average
For reading and writing females scored the most on average.***

```python
data.groupby('test preparation course').agg('mean')
```

***Inference-The students who completed the test preparation course have clearly got more marks on an average for maths,reading and writing.While the syudents who didn't attend any test preparation course got comparitively less marks on an average.***

***4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course***

```python
data.groupby('gender').var()
```

***For both male and female the highest scoring variation is evident for maths.It is followed by writing score.Reading score has the least variation in both categories.***

```python
data.groupby('test preparation course').var()
```

***The highest variation in scoring is for maths marks for students who attended the course as well for students who didnot attend any course.For students who completed the course the next highest variation is for reading followed by writing.For students who did not attend any course the next highest variation is for writing followed by reading.***
